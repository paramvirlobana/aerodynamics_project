```python
import math
import numpy
from matplotlib import pyplot
import pandas

%matplotlib inline
```

```python
naca4412_cp = pandas.read_csv(r"data\NACA4412_cp.csv")
x = naca4412_cp['X']
naca4412_cp_1 = naca4412_cp['c_p_1']
naca4412_cp_8 = naca4412_cp['c_p_8']
naca4412_cp_14 = naca4412_cp['c_p_14']


pyplot.figure(figsize=(10,6))

pyplot.ylim(-8.5, 1.5)


pyplot.plot(x, naca4412_cp_1, label = 'alpha = 1')
pyplot.plot(x, naca4412_cp_8, label = 'alpha = 8')
pyplot.plot(x, naca4412_cp_14, label = 'alpha = 14')
ax = pyplot.gca()
ax.invert_yaxis()
pyplot.title(r'$c_p$ Distribution - Finite Wing Theory')
pyplot.xlabel(r'$x/C$', fontsize=16)
pyplot.ylabel(r'$c_p$', fontsize=16)
pyplot.legend()
pyplot.grid(True, linestyle='--')
pyplot.savefig('c_p_xflr5.eps', format='eps', dpi = 1600)
```
